```python

#step 2 importing libraries
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM, AdamW
from datasets import load_dataset
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import numpy as np

# Step 3: Define LLMReasoningEvaluator Class
class LLMReasoningEvaluator:
    def __init__(self, model_name, task_type="causal", device="cuda" if torch.cuda.is_available() else "cpu"):
        self.device = device
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)

        # Set the padding token
        if not self.tokenizer.pad_token:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        if task_type == "causal":
            self.model = AutoModelForCausalLM.from_pretrained(model_name).to(self.device)
        elif task_type == "seq2seq":
            self.model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(self.device)
        else:
            raise ValueError("Invalid task_type. Use 'causal' or 'seq2seq'.")

    def train(self, dataset, epochs=1, learning_rate=5e-5):
        self.model.train()
        optimizer = AdamW(self.model.parameters(), lr=learning_rate)

        for epoch in range(epochs):
            print(f"Epoch {epoch+1}/{epochs}")
            total_loss = 0
            for idx, data in enumerate(dataset):
                # Skip examples with empty outputs
                if not data["output"].strip():
                    print(f"Skipping example {idx} due to empty output.")
                    continue

                # Tokenize input
                inputs = self.tokenizer(data["input"], return_tensors="pt", padding="max_length", truncation=True, max_length=512).to(self.device)
                # Tokenize output
                with self.tokenizer.as_target_tokenizer():
                    labels = self.tokenizer(data["output"], return_tensors="pt", padding="max_length", truncation=True, max_length=512).to(self.device)["input_ids"]

                # Replace padding token id in labels with -100 to ignore them in the loss computation
                labels[labels == self.tokenizer.pad_token_id] = -100

                # Ensure batch size alignment
                if inputs["input_ids"].size(0) != labels.size(0):
                    print(f"Skipping example {idx} due to batch size mismatch. Input size: {inputs['input_ids'].size(0)}, Label size: {labels.size(0)}")
                    continue

                # Debug shapes
                print(f"Example {idx} - Input shape: {inputs['input_ids'].shape}, Labels shape: {labels.shape}")

                # Forward pass
                outputs = self.model(**inputs, labels=labels)
                loss = outputs.loss

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                total_loss += loss.item()

            print(f"Loss: {total_loss / len(dataset)}")

    def predict(self, prompt):
        self.model.eval()
        with torch.no_grad():
            inputs = self.tokenizer(prompt, return_tensors="pt", padding=True, truncation=True, max_length=512).to(self.device)
            outputs = self.model.generate(**inputs, max_length=50)
        return self.tokenizer.decode(outputs[0], skip_special_tokens=True)

    @staticmethod
    def evaluate_metrics(predictions, references):
        bleu_scores = [sentence_bleu([ref.split()], pred.split()) for pred, ref in zip(predictions, references)]
        rouge = rouge_scorer.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)
        rouge_scores = [rouge.score(ref, pred) for pred, ref in zip(predictions, references)]
        f1 = f1_score([1 if ref == pred else 0 for pred, ref in zip(predictions, references)], [1] * len(references))
        return {
            "BLEU": sum(bleu_scores) / len(bleu_scores),
            "ROUGE-1": sum([score["rouge1"].fmeasure for score in rouge_scores]) / len(rouge_scores),
            "ROUGE-L": sum([score["rougeL"].fmeasure for score in rouge_scores]) / len(rouge_scores),
            "F1": f1
        }

    @staticmethod
    def visualize_results(predictions, references, title):
        correct = [1 if pred == ref else 0 for pred, ref in zip(predictions, references)]
        labels = [f"Example {i+1}" for i in range(len(predictions))]

        plt.figure(figsize=(10, 6))
        plt.bar(labels, correct, color=['green' if x else 'red' for x in correct])
        plt.xlabel("Examples")
        plt.ylabel("Correct Prediction (1 = Yes, 0 = No)")
        plt.title(title)
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()

# Step 4: Load and Prepare Datasets
print("Loading datasets...")

# Load a subset of the training dataset from CommonsenseQA
commonsense_dataset = load_dataset("commonsense_qa", split="train[:10%]")  # Using only 10% of the dataset
commonsense_training_data = []
for ex in commonsense_dataset:
    choices = ex["choices"]["text"]
    answer_key = ex["answerKey"]  # This is a string like "A", "B", etc.
    answer_index = ord(answer_key) - ord("A")
    commonsense_training_data.append({
        "input": ex["question"] + " " + " ".join(choices),
        "output": choices[answer_index]
    })

# Load a subset of the test dataset from Winograd Schema Challenge
wsc_dataset = load_dataset("winograd_wsc", "wsc285", split="test[:10%]", trust_remote_code=True)  # Using only 10% of the dataset
wsc_training_data = [
    {
        "input": ex["text"],
        "output": ex["options"][ex["label"]]
    }
    for ex in wsc_dataset
]

# Combine datasets
full_training_data = commonsense_training_data + wsc_training_data

print("Datasets loaded and prepared!")

# Step 5: Train and Evaluate Models
print("Initializing models...")

model1 = LLMReasoningEvaluator("gpt2", task_type="causal")
model2 = LLMReasoningEvaluator("t5-small", task_type="seq2seq")

print("Training Model 1 (GPT-2)...")
model1.train(full_training_data, epochs=3)

print("Training Model 2 (T5-small)...")
model2.train(full_training_data, epochs=3)

# Step 6: Evaluation (Optional)
print("Evaluating models...")

# Failure examples for CommonsenseQA
commonsenseqa_failure_examples = [
    ("Where do you put a letter after writing it? Choices: (a) Mailbox (b) Refrigerator (c) Oven", "Mailbox"),
    ("What do people wear on their head to see better? Choices: (a) Glasses (b) Shoes (c) Gloves", "Glasses")
]

predictions1_cqa = [model1.predict(prompt) for prompt, _ in commonsenseqa_failure_examples]
references_cqa = [ref for _, ref in commonsenseqa_failure_examples]

# Metrics for Model 1
metrics1_cqa = LLMReasoningEvaluator.evaluate_metrics(predictions1_cqa, references_cqa)
print("Model 1 (CommonsenseQA Failures):", metrics1_cqa)

# Visualization
LLMReasoningEvaluator.visualize_results(predictions1_cqa, references_cqa, "Model 1 - CommonsenseQA Predictions")


```
